```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve, average_precision_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout, BatchNormalization, Attention, GlobalAveragePooling1D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, TensorBoard
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler, RobustScaler
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
import os
import joblib
from datetime import datetime

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# 1. Enhanced Data Loading and Preparation with multiple improvements
def load_and_prepare_data(normal_path, abnormal_path, test_size=0.2, random_state=42):
    # Load data
    df_normal = pd.read_csv(normal_path, header=None)
    df_abnormal = pd.read_csv(abnormal_path, header=None)

    # Combine and shuffle
    df = pd.concat([df_normal, df_abnormal])
    df = df.sample(frac=1, random_state=random_state).reset_index(drop=True)

    # Extract features and labels
    X = df.iloc[:, :-1].values[:, :187]  # Use first 187 points
    y = df.iloc[:, -1].values

    # Robust scaling (better for outliers than StandardScaler)
    scaler = RobustScaler()
    X = scaler.fit_transform(X)

    # Reshape for CNN-LSTM
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y)

    # Apply SMOTE to handle class imbalance (only on training data)
    smote = SMOTE(random_state=random_state)
    X_train_resampled, y_train_resampled = smote.fit_resample(
        X_train.reshape(X_train.shape[0], -1), y_train)
    X_train_resampled = X_train_resampled.reshape((-1, X_train.shape[1], X_train.shape[2]))

    return X_train_resampled, X_test, y_train_resampled, y_test, scaler

# 2. Advanced Hybrid CNN-LSTM Model with Attention Mechanism
def create_enhanced_model(input_shape):
    model = Sequential()

    # First Conv Block
    model.add(Conv1D(128, 10, activation='relu', 
                    kernel_regularizer=l1_l2(l1=0.001, l2=0.001),
                    input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(3))
    model.add(Dropout(0.3))

    # Second Conv Block
    model.add(Conv1D(128, 7, activation='relu', 
                    kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(3))
    model.add(Dropout(0.3))

    # Third Conv Block
    model.add(Conv1D(64, 5, activation='relu',
                    kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(2))
    model.add(Dropout(0.3))

    # LSTM layers with Attention
    model.add(LSTM(128, return_sequences=True))
    model.add(Dropout(0.4))
    
    # Attention Mechanism
    model.add(GlobalAveragePooling1D())
    
    # Dense layers
    model.add(Dense(128, activation='relu', 
                   kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.4))
    
    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Custom optimizer with weight decay
    optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)

    model.compile(optimizer=optimizer,
                 loss='binary_crossentropy',
                 metrics=['accuracy', 
                          tf.keras.metrics.AUC(name='auc'),
                          tf.keras.metrics.Precision(name='precision'),
                          tf.keras.metrics.Recall(name='recall'),
                          tf.keras.metrics.TruePositives(name='tp'),
                          tf.keras.metrics.FalsePositives(name='fp'),
                          tf.keras.metrics.TrueNegatives(name='tn'),
                          tf.keras.metrics.FalseNegatives(name='fn')])
    return model

# 3. Enhanced Training with Class Weights and More Callbacks
def train_enhanced_model(model, X_train, y_train, epochs=100, batch_size=128, val_split=0.2):
    # Compute class weights for imbalanced data
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    class_weights = {i: class_weights[i] for i in range(len(class_weights))}
    
    # Create log directory for TensorBoard
    log_dir = "logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
    
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=7, min_lr=1e-6, verbose=1),
        ModelCheckpoint('best_model.keras', monitor='val_auc', mode='max', save_best_only=True, verbose=1),
        TensorBoard(log_dir=log_dir, histogram_freq=1)
    ]
    
    history = model.fit(X_train, y_train,
                       epochs=epochs,
                       batch_size=batch_size,
                       validation_split=val_split,
                       callbacks=callbacks,
                       class_weight=class_weights,
                       verbose=2)
    return history

# 4. Comprehensive Model Evaluation with Threshold Optimization
def evaluate_enhanced_model(model, X_test, y_test):
    # Get predicted probabilities
    y_pred_prob = model.predict(X_test).flatten()
    
    # Find optimal threshold
    precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)
    fscore = (2 * precision * recall) / (precision + recall + 1e-10)
    ix = np.argmax(fscore)
    optimal_threshold = thresholds[ix]
    print(f'Best Threshold={optimal_threshold:.3f}, F-Score={fscore[ix]:.3f}')
    
    # Predict with optimal threshold
    y_pred = (y_pred_prob > optimal_threshold).astype(int)
    
    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=['Normal', 'Abnormal']))
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Abnormal'], 
                yticklabels=['Normal', 'Abnormal'])
    plt.title("Confusion Matrix (Optimal Threshold)")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.3f})', color='darkorange', lw=2)
    plt.plot([0,1], [0,1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()
    
    # Precision-Recall Curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
    avg_precision = average_precision_score(y_test, y_pred_prob)
    plt.figure()
    plt.plot(recall, precision, label=f'PR curve (AP = {avg_precision:.3f})', color='blue', lw=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend(loc='upper right')
    plt.show()
    
    return y_pred, y_pred_prob, optimal_threshold

# 5. Enhanced Training History Visualization
def plot_enhanced_history(history):
    plt.figure(figsize=(18,6))
    
    # Accuracy plot
    plt.subplot(1,3,1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Loss plot
    plt.subplot(1,3,2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # AUC plot
    plt.subplot(1,3,3)
    plt.plot(history.history['auc'], label='Train AUC')
    plt.plot(history.history['val_auc'], label='Val AUC')
    plt.title('AUC')
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Paths to your data
    NORMAL_PATH = r"data/ptbdb_normal.csv"
    ABNORMAL_PATH = r"data/ptbdb_abnormal.csv"
    INPUT_SHAPE = (187, 1)
    EPOCHS = 100
    BATCH_SIZE = 128
    
    print("Loading and preparing data with enhancements...")
    X_train, X_test, y_train, y_test, scaler = load_and_prepare_data(NORMAL_PATH, ABNORMAL_PATH)
    
    print("\nClass distribution:")
    print(f"Training: Normal {sum(y_train==0)}, Abnormal {sum(y_train==1)}")
    print(f"Testing: Normal {sum(y_test==0)}, Abnormal {sum(y_test==1)}")
    
    print("\nCreating enhanced model...")
    model = create_enhanced_model(INPUT_SHAPE)
    model.summary()
    
    print("\nTraining model with enhancements...")
    history = train_enhanced_model(model, X_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE)
    
    print("\nEvaluating model with optimal threshold...")
    y_pred, y_pred_prob, optimal_threshold = evaluate_enhanced_model(model, X_test, y_test)
    
    plot_enhanced_history(history)
    
    # Save model and artifacts
    save_dir = r"C:/Users/HP/Desktop/upworkhealt/saved_models"
    os.makedirs(save_dir, exist_ok=True)
    
    # Save the entire model
    model.save(os.path.join(save_dir, "enhanced_hybrid_cnn_lstm_final.keras"))
    
    # Save scaler
    joblib.dump(scaler, os.path.join(save_dir, 'enhanced_scaler.save'))
    
    # Save optimal threshold
    with open(os.path.join(save_dir, 'optimal_threshold.txt'), 'w') as f:
        f.write(str(optimal_threshold))
    
    print(f"\nAll artifacts saved at {save_dir}!")
```
